```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img_breast = cv2.imread('./week3/breast.tif')
img_dft = cv2.imread('./week3/dft.tif')
img_moon = cv2.imread('./week3/moon.tif')
img_spine = cv2.imread('./week3/spine.tif')
```

```python
brightness = 180
result_dft = (img_dft.astype('int16') + brightness).clip(0, 255).astype('uint8')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(4, 6))
axes[0].imshow(img_dft, cmap='gray')
axes[0].set_title('Original')
axes[1].imshow(result_dft, cmap='gray')
axes[1].set_title('Brightness adjusted')
axes[0].axis('off')
axes[1].axis('off')
plt.tight_layout()
plt.show()
```

```python
img_breast_gray = cv2.cvtColor(img_breast, cv2.COLOR_BGR2GRAY)
img_breast_gray_negative = 1 - img_breast_gray
fig, axes = plt.subplots(1, 2, figsize=(4, 6))
axes[0].imshow(img_breast_gray, cmap='gray')
axes[0].set_title('Original')
axes[1].imshow(img_breast_gray_negative, cmap='gray')
axes[1].set_title('Negative')
axes[0].axis('off')
axes[1].axis('off')
plt.tight_layout()
plt.show()
```

```python
img_moon_gray = cv2.cvtColor(img_moon, cv2.COLOR_BGR2GRAY)
threshold = 255/4
print(img_moon_gray.shape)
fig, axes = plt.subplots(1, 2, figsize=(4, 6))
axes[0].imshow(img_moon_gray, cmap='gray')
axes[0].set_title('Original')
axes[1].imshow(img_moon_gray > threshold, cmap='gray')
axes[1].set_title('Threshold-ed')
axes[0].axis('off')
axes[1].axis('off')
plt.tight_layout()
plt.show()
```
