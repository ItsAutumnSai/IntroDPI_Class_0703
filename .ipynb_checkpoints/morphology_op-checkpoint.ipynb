```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('salt_pepper.jpeg', 0)
kernel = np.ones((2,2), np.uint8)
# closing
img_dilation = cv2.dilate(img, kernel, iterations = 1)
img_erosion = cv2.erode(img_dilation, kernel, iterations = 1)
# opening
img_erosion2 = cv2.erode(img_erosion, kernel, iterations = 1)
img_dilation2 = cv2.dilate(img_erosion2, kernel, iterations = 1)

plt.tight_layout()

plt.subplots(1,3, layout='constrained')
plt.subplot(1,3,1)
plt.title("Salt&Pepper");
plt.imshow(img, cmap='gray')
plt.axis('off')
plt.subplot(1,3,2)
plt.title("Closing (Salt left)");
plt.imshow(img_erosion,cmap='gray')
plt.axis('off')
plt.subplot(1,3,3)
plt.title("Followed by Opening");
plt.imshow(img_dilation2,cmap='gray')
plt.axis('off')
plt.show()
```
